```python
import pandas as pd

# Load dataset
df= pd.read_csv("/content/student performance.zip")

# view first few rows
print(df.head())

# 1. Check missing values
print("Missing values:\n", df.isnull().sum())

# 2. Fill missing values in 'math score' if any
df["math score"].fillna(df["math score"].mean(), inplace=True)

# 3. Remove duplicates
df.drop_duplicates(inplace=True)

# 4. Drop rows with any remaining missing values
df.dropna(inplace=True)

# 5. Print cleaned dataset shape
Print("Final dataset shape:",df.shape)
```

```python
import matplotlib.pyplot as plt

#Sample data
x=[1,2,3,4,5]
y=[5,10,15,20,25]

# Plotting
plt.plot(x,y)

#Labels and title
plt.xlabel("x-axis (Time)")
plt.ylabel("y-axis (Value)")
plt.title("Simple Line Chart")

# Show the chart
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import pandas as pd

# Sample dataset
data={
    'Experince':[1,2,3,4.5,5.5], # Added 5.5 to match the length of 'Salary'
    'Salary':[40000,50000,60000,70000,80000]
}
df=pd.DataFrame(data)

x=df[['Experince']] # Corrected 'Experience' to 'Experince'
y=df['Salary']

# Step 1: Split data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2, random_state=42) # Added random_state for reproducibility

# Step 2:Create model
model = LinearRegression()

# Step 3: Train
model.fit(x_train,y_train)

# Step 4: Predict
prediction=model.predict(x_test)

# Step 4: Evaluate
mse=mean_squared_error(y_test,prediction)
print("Mean Squared Error:",mse)
```
